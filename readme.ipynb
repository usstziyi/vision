# 单发多框目标检测模型
### 定义
单发：神经网络只需要对输入图像进行一次前向传播，就能直接预测出图像中所有目标的类别和位置。  
结构：基础网络块 + 几个多尺度特征块  
### 加载数据集  
数据集只有类别标签和边界框坐标。不能直接用来训练。
### 定义模型
- 多层特征图
- 基于特征图生成锚框
- 制造标签(因为数据集的标签不能直接用来训练)
    - 根据交并比（IoU），把真实框(gt_boxes)分配给锚框(anchors)
        - 真实类别(gt_classes)
        - 真实偏移量(gt_offsets)
        - 真实掩码(gt_masks)
- 模型预测锚框的类别(pred_classes)和偏移量(pred_offsets)
- 损失函数：分类损失+回归损失
- 优化器：SGD
- 模型训练好




### 预测阶段
- 生成锚框（Anchor Box）
- 预测
    - 预测锚框的类别(pred_classes)
    - 预测锚框的偏移量(pred_offsets)，得到预测框(pred_boxes)
    - 对预测框进行非极大值抑制(NMS:得分+IoU)，得到最终的预测结果
