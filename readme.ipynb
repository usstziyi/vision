# 目标检测模型

训练阶段：
- 根据CNN生成多尺度特征图(feature maps)
- 根据每个特征图，生成锚框（Anchor Box）
- 随机预测锚框的类别(pred_classes)和偏移量(pred_offsets)
- 根据交并比（IoU），把真实框(gt_boxes)分配给锚框(anchors)
    - 真实类别(gt_classes)
    - 真实偏移量(gt_offsets)
- 训练
    - 损失函数：分类损失+回归损失
    - 优化器：SGD

预测阶段
- 生成锚框（Anchor Box）
- 预测
    - 预测锚框的类别(pred_classes)
    - 预测锚框的偏移量(pred_offsets)，得到预测框(pred_boxes)
    - 对预测框进行非极大值抑制(NMS:得分+IoU)，得到最终的预测结果

单发多框模型  
单发：神经网络只需要对输入图像进行一次前向传播，就能直接预测出图像中所有目标的类别和位置。  
结构：基础网络块 + 几个多尺度特征块
